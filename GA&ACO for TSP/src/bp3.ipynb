```python
import time
import matplotlib as plt 
from AntColonyOptimization import AntColonyOptimization
from GeneticAlgorithm import GeneticAlgorithm
from Maze import Maze
from PathSpecification import PathSpecification
from TSPData import TSPData
import matplotlib.pyplot as plt 
# Construct the optimization objects
maze = Maze.create_maze("./../data/hard_maze.txt")
spec = PathSpecification.read_coordinates("./../data/hard_coordinates.txt")

for i in range(1,10000,200):
    no_ants = 600 #number of ants
    no_gen = 20 #number of generation
    q = i
    evap = 0.1
    alpha = 1
    beta = 5 
    convergence_criterion = 10
    prev_shortest_route_array = []
    aco = AntColonyOptimization(maze, no_ants, no_gen, q, evap, alpha, beta, convergence_criterion)
    prev_shortest_route = aco.find_shortest_route(spec, prev_shortest_route_array)
    x=i/10000
    plt.plot(range(len(prev_shortest_route_array)),prev_shortest_route_array, color=(x,0,1-x), alpha=0.8)

# Save starting time
start_time = int(round(time.time() * 1000))
plt.xlabel("generations")
plt.ylabel("length of shortest route")
plt.title("performance over different Q for hard")
plt.show()
```
